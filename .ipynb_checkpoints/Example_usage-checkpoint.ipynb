Since this is only an example, we load a processed csv file with longitude and latitude in it.

````python
from scripts.plot_many_traj import *
from scripts.loader import *
from pylab import rcParams
rcParams['figure.figsize'] = (16,10)

# We load a different file, that has been transformed already.
csv = "transformed.csv"
netcdf = "O_WCB_all_20160922_00.nc"
output_param = ["T"]
input_param = ["dzeta"]
output = "pics/traj"
trajectory = range(1, 21)

# Loading the data
res = load_output(csv, change_ref=False)
net_df, pollat, pollon = load_nc(inp=netcdf, get_pol=True)
````

````python
# Plot the derivatives
for input_param in input_param:
    for output_param in output_param:
        filename = (output + "_deriv_" + output_param
                    + "_" + input_param + ".png")
        plot_weather_deriv(df=res,
                           in_param=input_param,
                           out_param=output_param,
                           filename=filename)
````

````python
# Plot the netCDF data
for out_param in output_param:
    filename = output + "_netcdf_" + out_param + ".png"
    plot_traj_nc(df=net_df,
                 out_param=out_param,
                 filename=filename,
                 trajs=trajectory,
                 max_time=None)
````
