```python
with open("../mogonii_restart.job", "r") as f:
    prefix = None
    trajs = None
    for line in f:
        if "INPUT_FILENAME=" in line:
            s = line.split("traj_")
            prefix = s[1][:-9]
        elif "my_srun build/apps" in line:
            s = line.split("..")
            s = s[1].split("}")
            trajs = s[0]
        if trajs is not None:
            print("PREFIX={}".format(prefix))
            print('$my_parallel "$my_srun python ratio.py {{}} $PREFIX" ::: {{0..{}}} &'.format(trajs))
            print("")
            prefix = None
            trajs = None
```
